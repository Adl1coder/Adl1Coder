```python
import numpy as np
import matplotlib.pyplot as plt

# Resmi ykle
image = cv2.imread('/content/Ekran grnts 2024-03-17 101042.png')

# Histogram eitleme ilemi
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_image)

# Histogram hesaplama
hist = cv2.calcHist([equalized_image],[0],None,[256],[0,256])

# Histogram izimi
plt.figure(figsize=(8, 6))
plt.title('Histogram')
plt.xlabel('Piksel Deeri')
plt.ylabel('Piksel Says')
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Resmi ykle
image = cv2.imread('/content/Ekran grnts 2024-03-17 101042.png')

# Histogram eitleme ilemi
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_image)

# Histogramlar izdirme
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Orjinal Gri Resim')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Eitlenmi Gri Resim')

plt.show()
```
